```python
#Humidity effect on air quality
def assign_humidity(h):
    if h < 50:
        return 'Low humidity'
    elif h <= 80:
        return 'High humidity'
    else:
        return 'Very high humidity'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

humidity_pivot = pd.pivot_table(
    air_data,
    index='sensor.name',
    columns='humidity_level',
    values='pm2.5_atm',
    aggfunc=['mean','median','max','count']
)

humidity_pivot
```

```python
#Temperature effect on air quality
def assign_temperature(T):
    if T < 32:
        return 'Below freezing'
    elif T <= 50:
        return 'Cool'
    elif T <= 70:
        return 'Warm'
    else:
        return 'Hot'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

temperature_pivot = pd.pivot_table(
    air_data,
    index='sensor.name',
    columns='temperature_level',
    values='pm10.0_atm',
    aggfunc=['mean','median','max','count']
)

temperature_pivot
```

```python
#EPA AQI Unhealthy for Sensitive Groups corresponds to: Pollutant	USG range PM2.5	35.5  55.4, PM10	155  254 wont calculate AQI just use these cutoffs.

pm25_usg_events = air_data[(air_data['pm2.5_atm'] >= 35.5) & 
                           (air_data['pm2.5_atm'] <= 55.4)]

pm25_usg_events = pm25_usg_events.sort_values(by='date', ascending=True)

pm25_usg_events[['date','sensor.name','pm2.5_atm','temperature','humidity','sensor.altitude']].head(25)

pm25_usg_summary = pm25_usg_events.groupby('sensor.name').agg(
    first_day = ('date','min'),
    last_day = ('date','max'),
    days_count = ('date','nunique'),
    max_pm25 = ('pm2.5_atm','max')
).sort_values(by='days_count', ascending=False)

pm25_usg_summary
```

```python
#Summarize PM10 USG by location
pm10_usg_events = air_data[(air_data['pm10.0_atm'] >= 155) & 
                           (air_data['pm10.0_atm'] <= 254)]

pm10_usg_events = pm10_usg_events.sort_values(by='date', ascending=True)

pm10_usg_events[['date','sensor.name','pm10.0_atm','temperature','humidity','sensor.altitude']].head(25)

pm10_usg_summary = pm10_usg_events.groupby('sensor.name').agg(
    first_day = ('date','min'),
    last_day = ('date','max'),
    days_count = ('date','nunique'),
    max_pm10 = ('pm10.0_atm','max')
).sort_values(by='days_count', ascending=False)

pm10_usg_summary
```
